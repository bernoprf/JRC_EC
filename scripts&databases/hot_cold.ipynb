```python
### NEDC COLD ### 

#os.chdir(gas_coldhot)

dic_nedccold = {}
for file in glob.glob(gas_coldhot + "*NEDC*COLD*.xl*"):         # THIS CAN BE XLS OR XLSX, check
    try:
        print(file.split('\\')[-1])
        df = pd.read_excel(file, sheetname = 'Data', header = 0, parse_cols = "A:BC")
        df = df.dropna(axis=0, how='all')

        NOx = df['c_NOx_Dil_g_vec\n[grams]']
        NOx_sum = NOx.sum()*1000
        CO2 = df['c_CO2_Dil_g_vec\n[grams]']
        CO2_sum = CO2.sum()
        t = len(df.index)

        NOx_start = df['c_NOx_Dil_g_vec\n[grams]'][0:300]
        NOx_start_sum = NOx_start.sum()*1000
        CO2_start = df['c_CO2_Dil_g_vec\n[grams]'][0:300]
        CO2_start_sum = CO2_start.sum()
        t_start = 300

        NOx_end = df['c_NOx_Dil_g_vec\n[grams]'][300:len(df.index)]
        NOx_end_sum = NOx_end.sum()*1000
        CO2_end = df['c_CO2_Dil_g_vec\n[grams]'][300:len(df.index)]
        CO2_end_sum = CO2_end.sum()
        t_end = t - t_start
    except:
        #print(file.split('\\')[-1])
        df = pd.read_excel(file, sheetname=0, header=None)

        vals = df.values
        mux = pd.MultiIndex.from_arrays(df.ffill(1).values[:2, 1:], names=[None, 'Time'])   # trick to create the proper MultiIndex
        df = pd.DataFrame(df.values[2:, 1:], df.values[2:, 0], mux)                         # same here

        df.drop(df.index[[0,1]],inplace=True)                                               # remove the 2 first rows after the headers

        NOx = df['Diluted grams']['NOx']
        NOx_sum = NOx.sum()*1000
        CO2 = df['Diluted grams']['CO2']
        CO2_sum = CO2.sum()
        t = len(df.index)

        NOx_start = df['Diluted grams']['NOx'][0:300]
        NOx_start_sum = NOx_start.sum()*1000
        CO2_start = df['Diluted grams']['CO2'][0:300]
        CO2_start_sum = CO2_start.sum()
        t_start = 300

        NOx_end = df['Diluted grams']['NOx'][300:len(df.index)]
        NOx_end_sum = NOx_end.sum()*1000
        CO2_end = df['Diluted grams']['CO2'][300:len(df.index)]
        CO2_end_sum = CO2_end.sum()
        t_end = t - t_start
        
    dic_nedccold[file] = [NOx_sum,CO2_sum,t,
                          NOx_start_sum,CO2_start_sum,t_start,
                          NOx_end_sum,CO2_end_sum,t_end]            # save means in the dictionary
    

lab = pd.DataFrame()                                     # convert dic to dataframe
lab = lab.from_dict(dic_nedccold, orient='index')             # proper order
lab.columns = ['NOx','CO2','t','NOx_start','CO2_start','t_start','NOx_end','CO2_end','t_end']               # headers
lab_nedccold_g = lab

lab_nedccold_g['new index'] = [i.split('\\')[-2] for i in lab_nedccold_g.index.values]
lab_nedccold_g.set_index('new index')
```

```python
lab_nedccold_g=lab_nedccold_g.groupby(lab_nedccold_g['new index']).mean()
lab_nedccold_g
```
